### Exercise 1

Given an array of positive integers `nums`, return a list of all of the negative integers.<br>
Ex. 1<br>
nums = [1, 3, 5, 7, 8]<br>
Expected Output: [-1, -3, -5, -7, -8]<br>

Ex. 2<br>
nums = [100, 534, 32, 15, 77, 222, 788, 345, 75645, 22]<br>
Expected Output: [-100, -534, -32, -15, -77, -222, -788, -345, -75645, -22]<br>

```python
def negative_nums(num_list):
    return [-(abs(num)) for num in num_list] 
#     for num in num_list:
#         negative_list.
#     return negative_list
example1 = negative_nums([1, 3, 5, 7, 8])
example2 = negative_nums( [100, 534, 32, 15, 77, 222, 788, 345, 75645, 22])
print(f"{example1}\n{example2}")
```

### Exercise 3

Given a string `digits`, return a **string** of the digits + 1<br>

Ex. 1<br>
digits = '123'<br>
Expected Output: '124'<br>

Ex. 2<br>
digits = '99'<br>
Expected Output: '100'<br>

```python
# Given a string digits, return a string of the digits + 1


# 1. Input(s): string of digits
# 2. Output(s): string of digits + 1 
    # (not each individual digit + 1, just the next counting number from the integer version of the digit)
# 3a. Constraints: string contains only digits
# 3b. Edge cases: empty string, no digits in string, floats, fractions, roman numerals in string
# 4. Steps:
# a) define a function with one parameter: string
# b) convert the string to int
# c) add 1 to int version of string
# d) convert int back to string
# e) return string

# 5. Pseudocode
    # define a function with 1 parameter -> string
        # convert str to int
        # int + 1
        # convert int to str
        # return str

# 6. Code
def string_plus_one(digit_string):
    return str(int(digit_string) + 1)
```

```python
# Ex. 1
# digits = '123'
# Expected Output: '124'

def string_plus_one(digit_string):
    return str(int(digit_string) + 1)

string_plus_one('123')
```

```python
# Ex. 2
# digits = '99'
# Expected Output: '100'

def string_plus_one(digit_string):
    return str(int(digit_string) + 1)

string_plus_one('99')
```

```python
# What if the goal was to increase EACH digit in the string by 1?
# Given a string of digits, return a string with EACH of the digits + 1

# 1. Input(s): string of digits
# 2. Output(s): string of digits + 1 
    # (not each individual digit + 1, just the next counting number from the integer version of the digit)
# 3a. Constraints: string contains only digits
# 3b. Edge cases: empty string, no digits in string, floats, fractions, roman numerals in string
        # what to do if the digit is 9? return 0?
    
# 4. Steps:
# a) define a function with one parameter: string
# b) loop through digits in string
# c) convert the digit to int
# d) add 1 to int version of digit
# e) convert int back to string
# f) return string

# 5. Pseudocode
    # define a function with 1 parameter -> string
        # for digit in string
            # convert str to int
            # int + 1
            # convert int to str
        # return str

# 6. Code
def each_digit_plus_one(digit_string):
    dig_up_one_string = ""
    for digit in digit_string:
        if digit == 9:
            dig_up_one_string += digit.replace(digit, 0)
        else:
            dig_up_one_string += digit.replace(digit, str(int(digit) + 1))
    return (dig_up_one_string)

```

```python
# Ex. 3
# digits = '123'
# Expected Output: '234'
def each_digit_plus_one(digit_string):
    dig_up_one_string = ""
    for digit in digit_string:
        if digit == 9:
            dig_up_one_string += digit.replace(digit, '0')
        else:
            dig_up_one_string += digit.replace(digit, str(int(digit) + 1))
    return (dig_up_one_string)

each_digit_plus_one('123')
```
